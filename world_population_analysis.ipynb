# Population trends from 2000s to 2023 for each region

```python
years = [str(year) for year in range(2000, 2024)]

df_region_trend = df_reg.groupby('Region')[years].sum()

df_region_trend = df_region_trend.T

df_region_trend.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
for region in df_region_trend.columns:
    plt.plot(df_region_trend.index, df_region_trend[region], label=region)


plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Population Trends by Region (2000 - 2023)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)


plt.show()
```

### Trends in the 21st century
_East Asia & pacific has the highest population contribution._  
_North America contributes the least._  
_East Asia & pacific,South Asia, Sub-Saharan Africa and Middle East &North Africa have been incling._  
_Europe & central asia and South Asia seems to have a constant growth with no dramatic peaks._  

### Region with fastest population growth

```python
df_region_trend.pct_change().mean().sort_values(ascending = False)
```

_Sub-Saharan Africa has the fastest population growth._  
_The least growth is Europe & Central Asia._  

### Decline or Stagnancy?

```python
change = df_region_trend.iloc[-1] - df_region_trend.iloc[0]
change.sort_values(ascending = False)
```

_No population is declining so far. However North America and Europe&Central Asia are growing very slowly._  

### Regions with the highest population

```python
df_region_trend.loc['2023'].sort_values(ascending=False)
```

_East Asia & Pacific is the largest then South Asia_
